```python
import matplotlib.pyplot as plt
import numpy as np
record = open("./reward_record.txt", 'r')

reward = []
for line in record.readlines():
    l = line.split(' ')
    l[-1] = l[-1].replace('\n', '')
    if float(l[-1]) < 500:
        reward.append(float(l[-1]))
print(reward)
```

```python
x = np.arange(0, 3001)
plt.plot(reward, label='reward')
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.title("Reward & Episode")
plt.legend()
plt.show()
```
